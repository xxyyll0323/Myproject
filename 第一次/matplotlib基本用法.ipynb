## 基础绘图

### 折线图

```python
import numpy as np
import matplotlib.pyplot as plt
# 生成数据
x = np.linspace(0, 10, 100)  # 0到10的100个点
y = np.sin(x)

# 绘制折线图
plt.plot(x, y, color='red', linestyle='--', label='sin(x)')  # 颜色、线型、图例
plt.title("正弦曲线")  # 标题
plt.xlabel("X轴")      # X轴标签
plt.ylabel("Y轴")      # Y轴标签
plt.legend()           # 显示图例
plt.grid(True)         # 显示网格
plt.show()             # 显示图形
```

### 散点图

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.randn(50)
y = x + np.random.randn(50) * 0.1
plt.scatter(x, y, color='blue', marker='o', alpha=0.6)  # 透明度alpha
plt.show()
```

### 柱状图

```python
import numpy as np
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C']
values = [10, 25, 15]
plt.bar(labels, values, color=['#FF9999', '#66B2FF', '#99FF99'])  # 自定义颜色
plt.show()
```

### 直方图

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(1000)
plt.hist(data, bins=30, edgecolor='black')  # bins控制分箱数量
plt.show()
```

## 多子图绘制

```python
import numpy as np
import matplotlib.pyplot as plt
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # 2行2列，尺寸10x8英寸

# 子图1：折线图
axs[0,0].plot(x, y, color='green')
axs[0,0].set_title('子图1')

# 子图2：散点图
axs[0,1].scatter(x, y, marker='x')

# 子图3：柱状图
axs[1,0].bar(labels, values)

# 子图4：直方图
axs[1,1].hist(data, bins=20)

plt.tight_layout()  # 自动调整子图间距
plt.show()
```
