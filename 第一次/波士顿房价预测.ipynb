```python
import pandas as pd
import matplotlib.pyplot as plt

# 定义列名
column_names = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE',
    'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'
]

# 直接从 URL 读取数据
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'
df = pd.read_csv(data_url, delim_whitespace=True, names=column_names)
#使用 pandas 的 read_csv 函数从指定的 URL 读取数据，中间的表示数据以空格分隔

# 数据预处理
# 检查缺失值
missing_values = df.isnull().sum()
print("各列缺失值数量：")
print(missing_values)

# 分析各特征的描述性统计信息
statistics = df.describe()
print("各特征的描述性统计信息：")
print(statistics)

# 数据可视化
# 1. 房价（MEDV）分布直方图
plt.figure(figsize=(10, 6))  #创建一个大小为 10x6 英寸的图形窗口
plt.hist(df['MEDV'], bins=20, color='skyblue', edgecolor='black') 
# 绘制 MEDV 列（房价）的直方图，将数据分为 20 个区间，颜色为天蓝色，边框颜色为黑色
plt.title('波士顿房价分布直方图')
plt.xlabel('房价')
plt.xticks(rotation=45)
plt.ylabel('频数')
plt.show()

# 2. 部分特征与房价的散点图
features = ['RM', 'LSTAT', 'PTRATIO']  #定义了包含三个特征名称的列表
for feature in features:
    plt.figure(figsize=(10, 6))
    plt.scatter(df[feature], df['MEDV'], color='green')
    plt.title(f'{feature} 与 房价 的散点图')
    plt.xlabel(feature)
    plt.ylabel('房价')
    plt.show()


    
```
